```python
#importando bibliotecas
import pandas as pd
import requests as rq
import json as js
import numpy as np
```

```python
url =  'https://fakerapi.it/api/v1/persons?_quantity=300_&_gender=male&_birthday_start=2005-01-01'
```

```python
client = rq.get(url, verify = True).json()
df = pd.DataFrame(client)
df
```

```python
df_genero = pd.DataFrame(client['data'])
```

```python
df_genero
```

```python
#padronizando os generos
df_genero['gender'].replace(['male'], 'M', inplace=True)
df_genero['gender'].replace(['female'], 'F', inplace=True)
```
